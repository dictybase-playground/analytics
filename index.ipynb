# Python jupyter notebook

This notebook shows using a Python environment with [polars](https://pola-rs.github.io/polars-book/user-guide/introduction.html), including , all of which were specificied using a `conda` [environment.yml](../edit/environment.yml) file. 

## Importing python module

```python
import polars as pl
```

## Read data file(csv format) on covid-19

```python
df = pl.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
df.head()
```

## Selecting columns of interesting data

```python
ldf = df.select(["location","new_cases"])
ldf.head()
```

```python
cdf = ldf.groupby("location").agg([pl.col("new_cases").sum().alias("count")]).sort("count", reverse=True).limit(5)
cdf.head()
```
